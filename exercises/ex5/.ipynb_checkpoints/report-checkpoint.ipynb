### Part a
Using Weiner filter, try to recustruct the blurred `Khayyam.jpg` image. Assume that the cameras motion-sensor returned the `psf_1.bmp` as impulse respunse of destruction function.

```python
import cv2 as cv
import numpy as np 
import matplotlib.pyplot as plt
from scipy.fftpack import fft2, ifft2, ifftshift
%matplotlib inline

### 
# Hint: Dont forget to scale [0-1] and normalize psf
#       PSF image should be shifted before taking dft (H = fft2( ifftshif(psf) )
#       > Regards
###

def showImages(img):
    for i in range(len(img)):
        plt.subplot(1,len(img),i+1)
        plt.imshow(img[i], cmap='gray', vmin=0, vmax=255)
    plt.tight_layout()
    plt.show()

khayyam = cv.imread('./Khayyam.jpg', cv.IMREAD_GRAYSCALE)
psf1 = cv.imread('./psf1.bmp', cv.IMREAD_GRAYSCALE)
showImages([khayyam, psf1])
```

```python
def normalize(mat):
    mat.astype(np.float32)
    return mat/mat.sum()

def weiner(img, psf, k=0):
    psf = normalize(psf)
    psf = ifftshift(psf)
    H = fft2(psf)
    H_2 = (np.abs(H) * np.abs(H))
    F = fft2(img)
    F = np.divide(H_2,(H_2 + k))*F
    F = np.divide(F,H)
    f = ifft2(F)
    f = np.abs(f)
    return f

showImages([weiner(khayyam, psf1, 10e-4)])
```

### Part b
The picture `car.jpg` is taken from a moving car on the street and, we want to read its plate. using previously defined function in part 1, find the propper parameters for weiner filter to clearify the plate, you can use `cv.line()` function to draw you desired `PSF` file.

```python
car = cv.imread("car.jpg", cv.IMREAD_GRAYSCALE)
showImages([car])
```

```python
from math import cos, sin, pi, degrees
def line(r, teta):
    x = r*cos(teta)
    y = r*sin(teta)
    return (int(x),int(y))

h, w = car.shape
psf = np.zeros(car.shape)
x, y = line(10, (15*pi)/360)
cv.line(psf, (h//2-x, w//2+y), (h//2+x, w//2-y), 255, 1)
plate = weiner(car, psf, 8*10e-4)
showImages([plate])
```

## Question 2
Using the code bellow, calculate sobol and Canny filter of the image.

```python
import cv2 as cv
import numpy as np

def sobol(img):
    sobol = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
    img = cv.filter2D(img, -1, sobol)
    return img

def threshold(img,t):
    h,w = img.shape
    for i in range(h):
        for j in range(w):
            img[i,j] = 0 if (img[i,j] < t) else 255
    return img

cam_id = 0

cap = cv.VideoCapture(cam_id)

while True:
    ret, I = cap.read()
    cv.imshow("My Camera", I)
    gray = cv.cvtColor(I, cv.COLOR_BGR2GRAY)
    gray = cv.GaussianBlur(gray,(5,5),3)
    canny_I = cv.Canny(gray, 60, 120)
    sobol_I = threshold(sobol(gray), 80)
    cv.imshow("Canny", canny_I)
    cv.imshow("Sobol", sobol_I)

    # Press 'e' to exit
    if cv.waitKey(1) & 0xFF == ord('e'):
        break 

cap.release()
cv.destroyAllWindows()
```

```python
import cv2 as cv
import numpy as np
from math import cos, sin, tan, atan
from tqdm.notebook import tqdm

comb = cv.imread("comb.jpg")
showImages([comb])
```

```python
def Hough(img, teta, delta):
    h, w = img.shape
    H = np.zeros((max(w,h) ,180))
    for i in tqdm(range(h)):
        for j in range(w):
            for degree in range(min(teta-delta, teta+delta), max(teta-delta, teta+delta)):
                p = x*cos((teta/180)*pi) + y*sin((teta/180)*pi)
                p = int(p)
                H[p, degree-abs(delta)] += 1
    #which lines must displayed on result, large and local maximum
    res = np.zeros(H.shape)
    for i in tqdm(range(max(w,h))):
        for j in range(180):
            if(H[i,j] > np.max(H)-10 and H[i,j] <= np.max(H)):
                res[i,j] = 1
    return res
blur = cv.GaussianBlur(cv.cvtColor(comb, cv.COLOR_BGR2GRAY), (5,5),3)
canny = cv.Canny(blur, 60, 120)
houghLines = Hough(canny, 90, 90)
```

```python
h, w = canny.shape
for p in tqdm(range(houghLines.shape[0])):
    for teta in range(houghLines.shape[1]):
        if(houghLines[p, teta] == 1):
            for i in range(h):
                for j in range(w):
                    e = j*cos((teta/180)*pi) + i*sin((teta/180)*pi)
                    if(p-e <= 1e-1):
                        comb[i,j] = (255,0,0)
showImages([comb])
```

### Part b
Cout number of coins which could be seen in the picture `coin.jpg`.

```python
import cv2
import numpy as np

img = cv2.imread('coin.jpg',0)
img = cv2.medianBlur(img,5)
cimg = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR)

img = cv.Canny(img, 60, 150)

circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,1,30,
                            param1=50,param2=50,minRadius=0,maxRadius=0)

circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    # draw the outer circle
    cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)
    # draw the center of the circle
    cv2.circle(cimg,(i[0],i[1]),2,(0,0,255),3)
showImages([img])
showImages([cimg])
```
